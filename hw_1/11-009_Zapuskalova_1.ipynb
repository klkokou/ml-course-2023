
### Пиксельная маска части изображения

#### Дано:
- изображение (формат jpg/png)
- диапазон строк для маски (y_min, y_max)
- диапазон столбцов для маски (x_min, x_max)
- число разбиений на области (h, например, для h=10 маска разбивается на 10*10=100 равных областей)

```python
from skimage.io import imread
import matplotlib.pyplot as plt
# дано
params = {
    'h' : 9,
    'y_min' : 180, 
    'y_max' : 270,
    'x_min' : 280,
    'x_max' : 550}
orig_img = imread('https://cdnstatic.rg.ru/crop910x606/uploads/images/161/31/13/kinopoisk.ru-Shrek-13985.jpg')

plt.imshow(orig_img)
plt.show()
```

```python
def blur_shreks_eyes(pic, y_min, y_max, x_min, x_max, h):
    
    blurred_pic = pic.copy()

    ranges_x = range(x_min, x_max+1, (x_max - x_min)//h)
    ranges_y = range(y_max, y_min - 1, -1 *(y_max - y_min)//h)
    
    for i in range(len(ranges_y)-1):
        for j in range(len(ranges_x)-1):
            for k in range(0,3):
                blurred_pic[ranges_y[i+1]:ranges_y[i],ranges_x[j]:ranges_x[j+1],k]=orig_img[ranges_y[i+1]:ranges_y[i],ranges_x[j]:ranges_x[j+1],k].mean()
    return blurred_pic
```

```python
blurred_pic = blur_shreks_eyes(orig_img, **params)
plt.imshow(blurred_pic)
plt.show()
```
